***What is Broadcasting?***

Broadcasting allows NumPy to:

Perform operations on different-shaped arrays

Without writing loops

Automatically expand smaller arrays


```python
#without broadcasting
a = np.array([1, 2, 3])
b = np.array([10, 20])

#a + b   # ERROR
```

```python
a = np.array([1, 2, 3])
b = 10

a + b
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

arr + 10
```
